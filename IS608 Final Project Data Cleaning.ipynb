Some of the schools contain names that have a coma in them such as, Arthur Ashe, Jr. and Talbott, Jr. This is throwing off pandas and making it believe that there is an extra column in the data since the names in the data are not contained in quotes. I've elected to handle this by cleaning each file using the function clean_file(). It takes each row of the data and looks for the ", Jr." patern. It then replaces this with  " Jr." allowing there data to be loaded in. The for loop is used to go ahead and clean all of the data files at once. 

```python
def clean_file(file):
    f_in = open(file, 'r')
    f_out_name = "clean_%s" % (file)
    f_out = open(f_out_name, mode='w')
    for line in f_in:
        line_out = re.sub(r', Jr.', ' Jr.', line)
        f_out.write(line_out)
    f_out.close()

for f in filenames:
    clean_file(f)
```

Now that the files in the data/608 have been cleaned we now need to read in each file, select the columns taht are needed to filter the data. We will keep the summary row for each school and the summary row for the state for the test data in each file.  

```python
cols = ["SCHOOL_YEAR", "LEVEL_CODE", "DIV_NUM", "DIV_NAME", "SUBJECT", "TEST_LEVEL", "TEST", "GENDER", "FEDERAL_RACE_CODE",
        "DISABILITY_FLAG", "LEP_FLAG","DISADVANTAGED_FLAG", "AVG_SOL_SCALE_SCORE", "PASS_ADVANCED_RATE", "PASS_PROF_RATE",
        "PASS_RATE", "FAIL_RATE"]

def read_data(f_list, cols):
    df = []
    for f in f_list:
        data = pd.read_csv(f, usecols=cols) # Read the csv file
        data = data[(data.LEVEL_CODE == "DIV") | (data.LEVEL_CODE == "STATE")] # Filtering only the sate and division data
        data = data[(data.GENDER.isnull()) & (data.FEDERAL_RACE_CODE.isnull()) & 
                    (data.DISABILITY_FLAG.isnull()) & (data.LEP_FLAG.isnull()) & 
                    (data.DISADVANTAGED_FLAG.isnull()) & (data.TEST_LEVEL.isnull()) & 
                    (data.TEST.notnull()) & (data.AVG_SOL_SCALE_SCORE.notnull)] # Filter's out all rows except for the summary row
        df.append(data)
    final_df = pd.concat(df)
    return(final_df)
```
